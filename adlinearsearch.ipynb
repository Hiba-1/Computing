```python
"""
linear search using a function
adliearsearch
Hiba
"""
#defining a function
def linearSearch(list, n, x):

    # Going through list sequencially
    for i in range(0, n):
        if (list[i] == x):
            return i
    return -1


list = [2, 4, 0, 1, 9]
x = 1
n = len(list)
result = linearSearch(list, n, x)
if(result == -1):
    print("Element not found")
else:
    print("Element found at index: ", result)
```
